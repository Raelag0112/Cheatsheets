# 1. Pyplot #

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x, y, c = "red", lw = 5, ls = "--")
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.plot(x, y)
```

```python
plt.figure()

plt.plot(x, y, label = "quadratique")
plt.plot(x, x**3, label = "cubique")
plt.title("Figure 1")
plt.xlabel("axe x")
plt.ylabel("axe y")
plt.legend()

plt.show()

plt.savefig("figure.png")
```

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, y, c = "red")
plt.title("graphique 1")

plt.subplot(2, 1, 2)
plt.plot(x, np.sin(x), label = "sinus")
plt.plot(x, np.cos(x), label = "cosinus")
plt.legend()

plt.show()
```

### POO ###

```python
# fig est un objet
# ax est un tableau ndarray qui contient des objets

fig, ax = plt.subplots(2, 1, sharex = True)
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```

```python
dataset = {f"experience {i}" : np.random.randn(100, 3) for i in range(5)}
```

```python
def graphique(data):
    
    plt.figure(figsize = (12, 30))
    
    for k, i in zip(data.keys(), range(1, len(data) + 1)):
        plt.subplot(len(data), 1, i)
        plt.plot(data[k])
        plt.title(k)
        
    plt.show()
```

```python
graphique(dataset)
```

### Scatter classification ###

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f"x contient {x.shape[0]} exemples et {x.shape[1]} variables")
print(f"il y a {np.unique(y).size} classes")
y
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y, alpha = 0.5, s = x[:, 2]*100)
plt.xlabel("longeur sépal")
plt.ylabel("largeur sépal")
```

### Graphique 3D ###

```python
%matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection = "3d")
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c = y)
```

```python
f = lambda x, y : np.sin(x) + np.cos(x + y)
```

```python
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

```python
Z.shape
```

```python
%matplotlib

ax = plt.axes(projection = "3d")
ax.plot_surface(X, Y, Z, cmap = "plasma")
```

### Histogramme ###

```python
plt.hist(x[:, 0], bins = 20)
plt.hist(x[:, 1], bins = 20)
```

```python
plt.hist2d(x[:, 0], x[:, 1], cmap = "Blues")
plt.xlabel("longueur sépal")
plt.ylabel("largeur sépal")
plt.colorbar()
```

```python
from scipy import misc
face = misc.face(gray = True)
#plt.imshow(face, cmap = "gray")
plt.hist(face.ravel(), bins = 255)
plt.show()
```

### Contour plot ###

```python
f = lambda x, y : np.sin(x) + np.cos(x + y) * np.cos(x)

X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)

plt.contour(X, Y, Z, 20, colors = "black")
```

```python
plt.contourf(X, Y, Z, 20, cmap = "RdGy")
plt.colorbar()
```
